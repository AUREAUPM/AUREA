# Representación de una elipse con Manim y documentada con Jupyter Notebook

Se parte de su ecuación paramétrica y se genera una animación donde:
- Se dibujan los ejes
- Se representa la curva
- Un punto recorre la elipse

## Ecuación parametrica de la elipse

Una elipse centrada en el origen y alargada verticalmente se define como:
 - x = cos(t)
 - y = 2*sin(t)

con t en el intervalo [0, 2π]

```python
# Carga la extensión correcta de Manim
%load_ext manim
```

```python
%%manim -ql Elipse 

from manim import *
import numpy as np 

class Elipse(Scene): 
    def construct(self):
        # Titulo
        titulo = Text("Elipse", font_size=36).to_edge(UP)
        self.play(Write(titulo))
        # Crea ejes
        ejes = Axes()
        # Define la elipse
        a = 1
        b = 2
        curva = ParametricFunction(
            lambda t: ejes.c2p(a*np.cos(t), b*np.sin(t)),
            t_range=[0, 2*PI],
            color=YELLOW
        )
        # Crea punto y trayectoria
        punto = Dot(color = BLUE)
        punto.move_to(ejes.c2p(a,0))
        trayectoria = TracedPath(punto.get_center, stroke_color=GREEN)

        # Animación
        self.play(Create(ejes))
        self.play(Create(curva))
        self.add(trayectoria, punto)
        self.play(MoveAlongPath(punto, curva), run_time=2, rate_func=linear)
        self.wait()
```

## Explicación de algunos aspectos del código

 - 'np' librería matematica
 - 'c2p' transforma coordenadas matematicas a coordenadas de pantalla
 - 'self.play' ejecuta la animación
 - 'Create()' dibuja los objetos progresivamente
 - 'self.add' añade objetos sin animación
